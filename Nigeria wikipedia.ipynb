## Evidence database from Wikipedia

We needed to expand our evidence database so we used Wikipedia to extract some information related to Nigerian elections.

```python
import wikipediaapi
```

```python
topics = ["Nigeria", "Muhammadu Buhari", "Yemi Osinbajo", "Ahmed Lawan", "Femi Gbajabiamila", "Olukayode Ariwoola", "House of Representatives (Nigeria)", "Senate of Nigeria","National Assembly (Nigeria)", "Chief Justice of Nigeria",
         "Speaker of the House of Representatives of Nigeria", "President of the Senate of Nigeria", "Vice President of Nigeria", "President of Nigeria","Federal government of Nigeria", "Constitution of Nigeria", "Fourth Nigerian Republic"]

wiki_wiki = wikipediaapi.Wikipedia('en')

all_documents = []
for termino in topics:
    #print(f"Extrayendo {termino}")
    page_py = wiki_wiki.page(termino)
    text = page_py.text
    all_documents.append(text)
```

```python
all_documents
```

```python
import pandas as pd
data = pd.DataFrame(all_documents)
```

```python
data.to_csv('nigeria_information.csv')
```

## Some data preprocessing

```python
data = pd.read_csv('factcheckers.csv')
```

```python
data
```

```python
data['Label'].value_counts()[0:13]
```

```python
l=["FALSE"  ,                  
"False" ,                   
"Incorrect",                 
"MISLEADING",                
"FALSE-"      ,              
"Correct"      ,             
"TRUE"          ,            
"Unproven"       ,           
"Mostly correct"  ,          
"INSUFFICIENT-EVIDENCE",     
"Misleading"            ,    
"false"                  ,   
"misleading"]      
```

```python
for i in range(len(data)):
    if data['Label'][i] not in l:
        data.drop(i, axis=0, inplace=True)
    else:
        if data['Label'][i] == 'Correct' or data['Label'][i] =='TRUE' or data['Label'][i] =='Mostly correct':
            data['Label'][i] = 'True'
        elif data['Label'][i] == 'Unproven' or data['Label'][i] =='INSUFFICIENT-EVIDENCE':
            data['Label'][i] = 'Unproven'
        else: 
            data['Label'][i] = 'False'
        
data
```
