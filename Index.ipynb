## Creating the index using whoosh having the collection directory.

```python
import os
import json
import re
from whoosh.index import create_in
from whoosh.fields import Schema,TEXT,ID,DATETIME,NGRAM,BOOLEAN,STORED,KEYWORD,NUMERIC
from whoosh.analysis import StemmingAnalyzer
```

```python
def load_json(filename):
    with open(filename) as fh:
        d=json.load(fh)
    return d
```

```python
schema=Schema(path=ID(stored=True,
                      analyzer=StemmingAnalyzer()),
              
              claim=TEXT(stored=True,analyzer=StemmingAnalyzer()),
              evidences=TEXT(stored=True,analyzer=StemmingAnalyzer()),
              label=TEXT(stored=True,analyzer=StemmingAnalyzer()),
              keywords=KEYWORD(stored=True,analyzer=StemmingAnalyzer()),
              date=TEXT(stored=True))


ix=create_in('index',schema)
writer=ix.writer()
```

```python
dir1='collection'
```

```python
for dirname,subdirs,files in os.walk(dir1):
    
    
    if "MAIN" not in dirname:
    
        for filename in files:

            fullname=os.path.join(dirname,filename)
            print(fullname)
            
                
                
            d=load_json(fullname)
            keyw=','.join(d['keywords'])
            writer.add_document(path=fullname,claim=d['claim'],evidences=d['evidences'],label=d['label'],keywords=keyw,date=d['date'])
        

writer.commit() 
```
